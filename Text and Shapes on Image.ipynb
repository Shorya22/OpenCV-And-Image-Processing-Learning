```python
import cv2 as cv
import numpy as np
```

## Text Over Image:

```python
img= cv.imread("Sample_image.jpeg")
img= cv.resize(img,(500,500))
# Add text to the image
cv.putText(
    # The image to add text to
    img=img,
    # The text to add
    text="Sample_Image",
    # The coordinates (x, y) of the bottom-left corner of the text
    org=(5, 25),
    # The font to use
    fontFace=cv.FONT_HERSHEY_TRIPLEX,
    # The scale of the font
    fontScale=1,
    # The color of the text (red in BGR format)
    color=(0, 0, 255),
    # The thickness of the text
    thickness=2,
    # The line type (anti-aliased lines)
    lineType=cv.LINE_AA
)

cv.imshow("sample",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

## Line on Image:

```python
img= cv.imread("Sample_image.jpeg")
img= cv.resize(img,(500,500))

new_img= cv.line(
    img=img,
    pt1=(20,400),
    pt2=(480,400),
    color=(0,0,255),
    thickness=2,
    lineType=4
)

cv.imshow("sample",new_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

## Rectangle Box on Image:

```python
# Read an image from a file named "Sample_image.jpeg"
img = cv.imread("Sample_image.jpeg")

# Resize the image to a fixed size of 500x500 pixels
img = cv.resize(img, (500, 500))

# Draw a rectangle on the image
new_img = cv.rectangle(
    # The image on which to draw the rectangle
    img=img,
    # The coordinates of the top-left corner of the rectangle
    pt1=(5, 320),
    # The coordinates of the bottom-right corner of the rectangle
    pt2=(495, 400),
    # The color of the rectangle (red in this case)
    color=(0, 0, 255),
    # The thickness of the rectangle's border
    thickness=2,
    # The type of line to use for the rectangle's border
    lineType=4
)

# Display the modified image in a window named "sample"
cv.imshow("sample", new_img)

# Wait for a key press
cv.waitKey(0)

# Close all OpenCV windows
cv.destroyAllWindows()
```

## Circle on Image:

```python

# Read an image from a file named "Sample_image.jpeg"
img = cv.imread("Sample_image.jpeg")

# Resize the image to a fixed size of 500x500 pixels
img = cv.resize(img, (500, 500))

# Draw a circle on the image
new_img = cv.circle(
    # The image on which to draw the circle
    img=img,
    # The coordinates of the center of the circle
    center=(190, 280),
    # The radius of the circle
    radius=80,
    # The color of the circle (red in this case)
    color=(0, 0, 255),
    # The thickness of the circle's border
    thickness=2,
    # The type of line to use for the circle's border
    lineType=1,
    # The number of bits of radius value to shift to the left before drawing the circle
    shift=0
)

# Display the modified image in a window named "sample"
cv.imshow("sample", new_img)

# Wait for a key press
cv.waitKey(0)

# Close all OpenCV windows
cv.destroyAllWindows()
```

## Polygon on Image:

```python

# Read an image from a file named "Sample_image.jpeg"
img = cv.imread("Sample_image.jpeg")

# Resize the image to a fixed size of 500x500 pixels
img = cv.resize(img, (500, 500))

# Draw the polygon on the image
new_img = cv.polylines(
    # The image on which to draw the polygon
    img=img,
    # The points that make up the polygon
    pts=[np.array([[184, 184], [252, 249], [226, 358], [140, 358], [115, 250]])],
    # Whether the polygon is closed or not
    isClosed=True,
    # The color of the polygon (red in this case)
    color=(0, 0, 255),
    # The thickness of the polygon's border
    thickness=5
)

# Display the modified image in a window named "sample"
cv.imshow("sample", new_img)

# Wait for a key press
cv.waitKey(0)

# Close all OpenCV windows
cv.destroyAllWindows()
```
