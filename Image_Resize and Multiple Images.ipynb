```python
import cv2
import numpy as np
```

## Image Resize:

```python
img= cv2.imread(r"Sample_image_2.jpeg") # Load Large size image
resize_img= cv2.resize(img,(800,500)) # 1st is image, and 2nd is w,h of image
cv2.imshow("Sample",resize_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Multiple Images:

```python
img= cv2.imread(r"Sample_image_2.jpeg") # Load Large size image
resize_img= cv2.resize(img,(300,300)) # 1st is image, and 2nd is w,h of image
multiple_images_h= np.hstack((resize_img,resize_img,resize_img))
multiple_images_v= np.vstack((multiple_images_h,multiple_images_h))
cv2.imshow("Multiple",multiple_images_v)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
