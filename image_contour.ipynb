# Real time Contour:

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    img = cv2.resize(frame, (600, 500))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    _, binary = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)
    m = np.array([[2,2], [2, 2]], np.uint8)  # corrected kernel
    binary = cv2.morphologyEx(binary, kernel=m, iterations=1, op=cv2.MORPH_OPEN)
    contours, _ = cv2.findContours(image=binary, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)
    new_img = cv2.drawContours(img, contours, -1, (0, 0, 255), 2)

    cv2.imshow("sample", new_img)
    if cv2.waitKey(25) & 0xff == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

# Contour Moment/Centre:

```python
import cv2
import numpy as np
```

```python
img= cv2.imread("contour.png")
img= cv2.resize(img,(600,500))
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,thr= cv2.threshold(gray,95,255,cv2.THRESH_BINARY)
cnt,_= cv2.findContours(thr,mode=cv2.RETR_TREE,method=cv2.CHAIN_APPROX_NONE)

for c in cnt:
    cv2.drawContours(img,cnt,-1,(0,0,255),2)
    m=cv2.moments(c)
    if m["m00"] != 0:
        x= int(m["m10"]/m["m00"])
        y= int(m["m01"]/m["m00"])
        cv2.circle(img,(x,y),2,(255,0,0),2)

cv2.imshow("org",img)
cv2.imshow("gray",gray)
cv2.imshow("binary",thr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Contour Area:

```python
img= cv2.imread("contour.png")
img= cv2.resize(img,(600,500))
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,binary= cv2.threshold(gray,90,255,cv2.THRESH_BINARY)

cnt,_= cv2.findContours(binary,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)

area_cnt=[]

for c in cnt:
    cv2.drawContours(img,c,-1,(0,0,255),2)
    ar= cv2.contourArea(c)
    area_cnt.append(ar)

cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Convex Hull:

```python
img= cv2.imread("sample_img7.jpg")
img= cv2.resize(img,(600,500))
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,binary= cv2.threshold(gray,90,255,cv2.THRESH_BINARY)

cnt,_= cv2.findContours(binary,mode=cv2.RETR_TREE,method=cv2.CHAIN_APPROX_NONE)

hull=[]

for c in cnt:
    h= cv2.convexHull(c)
    hull.append(h)

cv2.drawContours(img,hull,-1,(0,0,255),2)

cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
