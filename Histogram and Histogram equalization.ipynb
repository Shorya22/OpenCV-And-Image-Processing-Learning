```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Histogram Equalization

```python
img = cv2.imread("sample_img7.jpg")
img = cv2.resize(img, (600, 500))

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# hist= cv2.calcHist([gray],[0],None,[255],[0,255])   
# plt.plot(hist)
# plt.show()

new_img= cv2.equalizeHist(gray)

hist= cv2.calcHist([new_img],[0],None,[255],[0,255])   
plt.plot(hist)
plt.show()

cv2.imshow("sample",new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("sample_img7.jpg")
img = cv2.resize(img, (600, 500))

b, g, r = cv2.split(img)

b_eq = cv2.equalizeHist(b)
g_eq = cv2.equalizeHist(g)
r_eq = cv2.equalizeHist(r)

new_img = cv2.merge([b_eq, g_eq, r_eq])

cv2.imshow("sample", new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## CLAHE Histogram EQ:

```python
img = cv2.imread("sample_img7.jpg")
img = cv2.resize(img, (600, 500))

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# new_img= cv2.equalizeHist(gray)

hist= cv2.calcHist([gray],[0],None,[255],[0,255])   
plt.plot(hist)
plt.show()

cl= cv2.createCLAHE(clipLimit=3,tileGridSize=(32,32))
new_img= cl.apply(gray)

hist= cv2.calcHist([new_img],[0],None,[255],[0,255])   
plt.plot(hist)
plt.show()

cv2.imshow("sample1",gray)
cv2.imshow("sample",new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
