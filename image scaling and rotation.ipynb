```python
import cv2
import numpy as np
```

```python
img= cv2.imread("Sample_image.jpeg")

res_img= cv2.resize(img,(500,500))

cv2.imshow("res_image",res_img)
cv2.imshow("sample",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img= cv2.imread("Sample_image.jpeg")
res_img= cv2.resize(img,(400,400))

w,h=res_img.shape[0],res_img.shape[1]
m=cv2.getRotationMatrix2D(center=(w/2,h/2),angle=60,scale=1)
new_img= cv2.warpAffine(res_img,M=m,dsize=(w,h))

h= np.hstack((res_img,new_img))
cv2.imshow("sample",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
