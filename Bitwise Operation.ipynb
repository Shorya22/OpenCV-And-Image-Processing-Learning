```python
import cv2
import numpy as np
```

## convert any image into binary image:

```python
# Read the 3-channel image
img = cv2.imread('sample_image_3.jpeg')
img = cv2.resize(img,(500,500))

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply thresholding to convert to binary image
_, binary = cv2.threshold(gray,150, 255, cv2.THRESH_BINARY)

# Display the binary image
cv2.imshow('Binary', binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Bitwise Operations:

## Bitwise AND:

```python
img1= cv2.imread("imgb1.png")
img1= cv2.resize(img1,(300,300))
img2= cv2.imread("imgb2.png")
img2= cv2.resize(img2,(300,300))


new_img= cv2.bitwise_and(img1,img2)
h= np.hstack((img1,img2,new_img))

cv2.imshow("sample",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Bitwise OR:

```python
img1= cv2.imread("imgb1.png")
img1= cv2.resize(img1,(300,300))
img2= cv2.imread("imgb2.png")
img2= cv2.resize(img2,(300,300))


new_img= cv2.bitwise_or(img1,img2)
h= np.hstack((img1,img2,new_img))

cv2.imshow("sample",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Bitwise XOR:

```python
img1= cv2.imread("imgb1.png")
img1= cv2.resize(img1,(300,300))
img2= cv2.imread("imgb2.png")
img2= cv2.resize(img2,(300,300))


new_img= cv2.bitwise_xor(img1,img2)
h= np.hstack((img1,img2,new_img))

cv2.imshow("sample",h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Bitwise NOT:

```python
img1= cv2.imread("imgb1.png")
img1= cv2.resize(img1,(300,300))
img2= cv2.imread("imgb2.png")
img2= cv2.resize(img2,(300,300))


new_img1= cv2.bitwise_not(img1)
new_img2= cv2.bitwise_not(img2)
h1= np.hstack((img1,new_img1))
h2= np.hstack((img2,new_img2))
v= np.vstack((h1,h2))

cv2.imshow("sample",v)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
